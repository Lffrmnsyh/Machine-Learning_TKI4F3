# 4. Visualisasi Data

```python
# List fitur yang akan diubah menjadi numerik
fitur_numerik = ['Total experience', 
                 'Yearly brutto salary', 
                 'Yearly bonus and stocks', 
                 'Kurzarbeit hours per week', 
                 'Vacation days']

# Mengonversi setiap fitur menjadi numerik
for fitur in fitur_numerik:
    df[fitur] = pd.to_numeric(df[fitur], errors='coerce')
```

```python
for feature in fitur_numerik:
    plt.figure(figsize=(8, 6))
    df[feature].plot(kind='hist', bins=100, color='skyblue', edgecolor='black')
    plt.title(f'Histogram of {feature}')
    plt.xlabel(feature)
    plt.ylabel('Frequency')
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()
```

```python
# Menghitung korelasi antar fitur numerik
correlation_matrix = df# Memilih hanya kolom-kolom numerik
numeric_columns = df.select_dtypes(include=['float64', 'int64'])

# Menghitung korelasi antar fitur numerik
correlation_matrix = numeric_columns.corr()

# Menampilkan heatmap korelasi
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", annot_kws={"size": 10})
plt.title('Korelasi antar Fitur')
plt.show().corr()
```
