# 1. Memuat Library

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# 2. Memuat Data

```python
df = pd.read_csv('/kaggle/input/vehicle-sales-data/car_prices.csv')
df.head()
```

# 3. Feature Statistic

```python
# Menampilkan statistik deskriptif untuk fitur numerik
numeric_stats = df.describe()

# Menampilkan statistik deskriptif untuk fitur kategorikal
categorical_stats = df.describe(include=['object'])

print("Statistik Fitur Numerik:")
print(numeric_stats)

print("\nStatistik Fitur Kategorikal:")
print(categorical_stats)
```

# 4. Box Plot

```python
df.info()
```

```python
# Memilih fitur-fitur numerik
numeric_features = ['year', 'condition', 'odometer', 'mmr', 'sellingprice']

# Membuat box plot untuk setiap fitur numerik
for feature in numeric_features:
    plt.figure(figsize=(8, 6))
    sns.boxplot(x=df[feature])
    plt.title(f'Box Plot dari {feature}')
    plt.xlabel(feature)
    plt.show()
```

# 5. Scatter Plot

```python
# Menonaktifkan pesan peringatan
import warnings
warnings.filterwarnings("ignore")

# Memilih fitur-fitur numerik
numeric_features = ['year', 'condition', 'odometer', 'mmr', 'sellingprice']

# Membuat scatter plot untuk semua fitur numerik
sns.pairplot(df[numeric_features])
plt.show()
```

# 6. Heat Map

```python
# Memilih fitur-fitur numerik
numeric_features = ['year', 'condition', 'odometer', 'mmr', 'sellingprice']

# Menghitung korelasi antar fitur numerik
correlation_matrix = df[numeric_features].corr()

# Menampilkan heatmap untuk korelasi antar fitur numerik
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Heatmap Korelasi Fitur Numerik')
plt.show()
```

# 7. Histogram

```python
# Memilih fitur-fitur numerik
numeric_features = ['year', 'condition', 'odometer', 'mmr', 'sellingprice']

# Mengelompokkan fitur numerik ke dalam interval
for feature in numeric_features:
    plt.figure(figsize=(8, 6))
    sns.histplot(df[feature], bins=20, kde=False)
    plt.title(f'Histogram dari {feature}')
    plt.xlabel(feature)
    plt.ylabel('Frequency')
    plt.show()
```

# 8. Pie Chart

```python
# Memilih fitur-fitur numerik
numeric_features = ['year', 'condition', 'odometer', 'mmr', 'sellingprice']

# Mengelompokkan fitur numerik ke dalam interval
# (Contoh: kita akan mengelompokkan tahun menjadi interval 5 tahun)
df['year_group'] = pd.cut(df['year'], bins=range(2000, 2026, 5), right=False)

# Menghitung frekuensi masing-masing kelompok
year_group_counts = df['year_group'].value_counts()

# Membuat pie chart dari frekuensi kelompok tahun
plt.figure(figsize=(8, 6))
plt.pie(year_group_counts, labels=year_group_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Pie Chart dari Kelompok Tahun')
plt.axis('equal')  # Untuk memastikan lingkaran terlihat proporsional
plt.show()
```

# 9. Line Chart

```python
# Memilih fitur-fitur numerik
numeric_features = ['year', 'condition', 'odometer', 'mmr', 'sellingprice']

# Pilih fitur numerik untuk sumbu x dan y (misal: tahun dan harga jual)
x_feature = 'year'
y_feature = 'sellingprice'

# Mengurutkan data berdasarkan fitur sumbu x (tahun)
df_sorted = df.sort_values(by=x_feature)

# Membuat line chart
plt.figure(figsize=(10, 6))
plt.plot(df_sorted[x_feature], df_sorted[y_feature], marker='o', linestyle='-')
plt.title(f'Line Chart dari {y_feature} terhadap {x_feature}')
plt.xlabel(x_feature)
plt.ylabel(y_feature)
plt.grid(True)
plt.show()
```

# 10. Violin Plot

```python
# Memilih fitur-fitur numerik
numeric_features = ['year', 'condition', 'odometer', 'mmr', 'sellingprice']

# Membuat violin plot untuk setiap fitur numerik
plt.figure(figsize=(12, 8))
for i, feature in enumerate(numeric_features):
    plt.subplot(2, 3, i + 1)
    sns.violinplot(x=df[feature])
    plt.title(f'Violin Plot dari {feature}')
    plt.xlabel(feature)

plt.tight_layout()
plt.show()
```
