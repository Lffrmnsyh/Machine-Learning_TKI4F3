# 1. Memuat Library

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# 2. Memuat Data

```python
df = pd.read_csv('/kaggle/input/diamonds/diamonds.csv')
df.head()
```

# 3. Feature Statistic

```python
# Menampilkan statistik fitur
print("Statistik untuk setiap fitur:")
print(df.describe(include='all'))
```

# 4. Box Plot

```python
df.info()
```

```python
# Memilih fitur-fitur numerik
numeric_features = ['carat', 'depth', 'table', 'price', 'x', 'y', 'z']

# Membuat box plot
for feature in numeric_features:
    plt.figure(figsize=(8, 6))
    sns.set_palette("pastel")
    sns.boxplot(x=feature, data=df)
    plt.title(f'Box Plot for {feature}')
    plt.show()
```

# 5. Scatter Plot

```python
# Memilih fitur-fitur numerik
numerical_features = ['carat', 'depth', 'table', 'price', 'x', 'y', 'z']

# Membuat scatter plot untuk semua kombinasi fitur numerik
sns.pairplot(df[numerical_features])
plt.show()
```

# 6. Heat Map

```python
# Memilih fitur-fitur numerik
numerical_features = ['carat', 'depth', 'table', 'price', 'x', 'y', 'z']

# Menghitung matriks korelasi antar fitur numerik
correlation_matrix = df[numerical_features].corr()

# Membuat heatmap dari matriks korelasi
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Heatmap Korelasi antar Fitur Numerik')
plt.show()
```

# 7. Histogram

```python
# Memilih fitur-fitur numerik
numerical_features = ['carat', 'depth', 'table', 'price', 'x', 'y', 'z']

# Membuat histogram untuk setiap fitur numerik
for feature in numerical_features:
    plt.figure(figsize=(8, 6))
    df[feature].plot(kind='hist', bins=30, color='skyblue', edgecolor='black')
    plt.title(f'Histogram of {feature}')
    plt.xlabel(feature)
    plt.ylabel('Frequency')
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()
```

# 8. Pie Chart

```python
# Memilih fitur-fitur numerik
numerical_features = ['carat', 'depth', 'table', 'price', 'x', 'y', 'z']

# Menentukan batas kategori
bins = [0, 0.5, 1, 1.5, 2, 2.5, 3]

# Membuat kategori untuk setiap fitur numerik
for feature in numerical_features:
    plt.figure(figsize=(8, 6))
    categories = pd.cut(df[feature], bins=bins).value_counts()
    plt.pie(categories, labels=categories.index, autopct='%1.1f%%', startangle=140)
    plt.title(f'Pie Chart of {feature}')
    plt.axis('equal')  # Memastikan lingkaran terlihat sebagai lingkaran
    plt.show()
```
